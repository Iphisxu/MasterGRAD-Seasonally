```python
import os
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt

import sys
sys.path.append('../../../src/')
from namelist import *

# silence the warning note
import warnings
warnings.filterwarnings("ignore")
```

O3

```python
var = 'O3'

years = [2014, 2015, 2016, 2019, 2021, 2022]
df = pd.DataFrame(
    columns=years,
)
for year in years:
    df_city = pd.read_excel(obsSep + f'city_{var}_{year}.xlsx',index_col=0)
    df_site = pd.read_excel(obsSep + f'site_{var}_{year}.xlsx',index_col=0)
    avg_city = df_city.mean(axis=1,skipna=True).values
    avg_site = df_site.mean(axis=1,skipna=True).values
    df[year] = avg_city - avg_site

df.plot()
plt.show()
```
